biblioteca voltada para machine learning

gerando massa de dados:

gerando valores entre os planos x e y, com 200 individuos, fator de caracteristicas = 1, ruído = 30

```python
from sklearn.datasets import make_regression

x, y = make_regression(n_samples=200, n_features=1, noise=30)
```

agora com a biblioteca matplotlib criaremos um gráfico

```python
import matplotlib.pyplot as plt

# criando no gráfico
plt.scatter(x,y)

# mostrando o gráfico
plt.show()
```

iniciando a criação do modelo de ML

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

com o modelo criado, agora ele está pronto para receber os dados que darão origem ao modelo. Para apresentar os dados ao modelo usamos o método fit(). O método predict() indica que queremos aplicar a previsão nos valores de x. O resultado será uma previsão de y para cada valor de x apresentado.

```python
# pesquisar na documentção todos os argumentos que podem ser passados em fit() para apresentar os dados ao modelo
modelo.fit(x, y)
```

```python
modelo.predict(x)
```

mostra uma reta de predição com a média dos dados apresentados

```python
# criando no gráfico
plt.scatter(x,y)

plt.plot(x, modelo.predict(x), color='red', linewidth=3)
# mostrando o gráfico
plt.show()
```
